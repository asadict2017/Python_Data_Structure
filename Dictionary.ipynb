```python
#unordered (no integer position) Key value pair within curly braces and separted by comma where key:value separated by :
#data get operation
#Mutability
#Add operation
#delete
#creating dictionary using for loop from two lists
#Dictionary comprehension
#Loop on dictionary items (complex scenario)
#dictionary within dictionary

##-------------------><--------------------------------------------------------------------------------------------------------------
#Key value pair within curly braces and separted by comma where key must be unique
#empty dictionary:
empty_dict={}  # dict()
students={'name':'Asad','age':39,'sex':'Male'}
#data get operation using key within bracket
students['name']
#if key does not exists it will raise error and resolve the issue using the get()method that will return None if data not exists
print(students.get('name1'))
print(students.get('name1',"not available"))


#Mutability in place operation(modify existing object on fly)
#Add operation using new key and value
students['country']='Bangladesh'
students

#delete using del key word before the key index
del students['country']
students

#iterating over dictionary items using .Items()method it will convert dictionary key, value pair to tuple within the list
## number of comma separted variables =tuple(item,.,.)
students.items() # dict_items([('name', 'Asad'), ('age', '39'), ('sex', 'Male')])
for var1,var2 in students.items():
  print(var1,var2)
```

```python
#shallow copy
students
students_copy=students # it uses same memory location
print(students)
print(students_copy)
students["name"]="krish"
print(students)
print(students_copy)
#to maintain the students_copy original data the shallow copy method that uses the different memory location
students_copy2=students.copy()
students["name"]="krish2"
print(students)
print(students_copy2)
```

```python
#Iterating over dictionary
for key in students.keys():
  print(key)
#values
for val in students.values():
  print(val)
  #items
for key,val in students.items():
  print(key,val)
```

```python
#nested dictionary
students={"student1":{"name":'asad',"age":39},"student2":{"name":'krish',"age":32}}
students
#accessing nested dictionary
print(students["student1"]["name"])

#iterating over nested dictionary
for std_id, std_info in students.items():
  print(f"{std_id} {std_info}")
  for key,val in std_info.items():
     print(f"{key} {val}")
```

```python
#creating dictionary using for loop from two lists

subjects=['Math','Science','Geography']
Score=[80,75,90]
#output should {'Math':80,'Science':75,'Geography':90}

#creating the empty dictionary
sub_score={}

len_sub=len(subjects) #3
for i in range(len_sub):
  sub_score[subjects[i]]=Score[i]
sub_score



```

```python
#Dictionary comprehension of two lists
#syntax             { Key:Value for loop}   using zip method

subjects=['Math','Science','Geography']
Score=[80,75,90]

list(zip(subjects,Score)) # [('Math', 80), ('Science', 75), ('Geography', 90)]

sub_score={key:value for key,value in  zip(subjects,Score) }
print(sub_score) # {'Math': 80, 'Science': 75, 'Geography': 90}
```

```python
#Loop on dictionary items (complex scenario)

fav_foods={'asad':['sharma','Tehari'],'babu':['rice','polao','fish curry']}
#list(fav_foods.items()) ## Item() keep the dictionary data into tuple  [('asad', ['sharma', 'Tehari']), ('babu', ['rice', 'polao'])]
for friend,foods in fav_foods.items():
  print(f"\n{friend}'s favorite food:")
  for food in foods:
    print(food)

```

```python
#practical example considering the following list item use the dictionary for counting the frequency of items of list
list_items=[1,1,1,1,2,2,3,3,3,3,3,3,3,3,3,4,5]
frequency={}
for item in list_items:
  if item in frequency:
    frequency[item]=frequency[item]+1
  else:
    frequency[item]=1
print(frequency)
```
