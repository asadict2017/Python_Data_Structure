```python
#unpacking with star
numbers=(1,2,3,45,6,7,8,9,10)
first,*second,third=numbers
print(first)
print(second)  #star will get value after first and before the lase
print(third)
```

```python
#Nested tuple-->tuples within tuple
#similar to list
nested_tuple=((1,2,3),('a','b','c'))
#accessing tuple
#suppose get c
get_c=nested_tuple[1][2]
print(get_c)

#unpacking ways
tup1,tup2=nested_tuple
get_c=tup2[2]
print(get_c)

#iterating over nested tuple
for sub_tup in nested_tuple:
  for tup in sub_tup:
    print(tup,end=",")
  print()
```
